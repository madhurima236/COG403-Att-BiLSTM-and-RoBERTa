```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_labels(d1, d2, d3, d4, x_label, y_label, title):
    
    fig, ax = plt.subplots(figsize=(9, 5))
    index = np.arange(len(d1))
    bar_width = 0.15
    opacity = 0.8

    plt.bar(index - bar_width, d1, bar_width, alpha=opacity, color='darkmagenta', label='Att-BiLSTM')
    plt.bar(index, d2, bar_width, alpha=opacity, color='lightpink', label='Att-BiLSTM trained on 3 datasets')
    plt.bar(index + bar_width, d3, bar_width, alpha=opacity, color='darkcyan', label='RoBERTa')
    plt.bar(index + 2*bar_width, d4, bar_width, alpha=opacity, color='cyan', label='RoBERTa trained on 3 datasets')

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(index + bar_width, ('anger', 'fear', 'joy', 'love', 'sadness', 'surprise'))
    # position the legend inside the plot
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1, 0, 0.0), fancybox=True, shadow=True, ncol=4)
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns

def plot_heatmap(matrices, title):
    # Plot 4 heatmaps in one figure with same scale

    matrices = [np.array(matrix) for matrix in matrices]

    vmin = min([matrix.min() for matrix in matrices])
    vmax = max([matrix.max() for matrix in matrices])
    
    labels = ['Anger', 'Fear', 'Joy', 'Love', 'Sadness', 'Surprise']
    matrix_name = ["Att-BiLSTM", "Att-BiLSTM trained on 3 datasets", "RoBERTa", 'RoBERTa trained on 3 datasets']
    
    fig, axs = plt.subplots(2, 2, figsize=(9, 9))
    # fig.suptitle(title, fontsize=16)
    fig.tight_layout(pad=3.0)

    for i, matrix in enumerate(matrices):
        row = i // 2
        col = i % 2
        ax = axs[row][col]
        ax.set_title(matrix_name[i])
        sns.heatmap(matrix, ax=ax, cmap="Reds", annot=True, fmt="d", cbar=False, square=True, linewidths=0.5, linecolor="black", vmin=vmin, vmax=vmax, xticklabels=labels, yticklabels=labels)

    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(axs[0][0].collections[0], cax=cbar_ax)

    # label y and x axis of the figure and bold the text
    fig.text(0.5, 0.0, 'Predicted Emotion', ha='center', va='center', fontsize=14, fontweight='bold')
    fig.text(0.0, 0.5, 'True Emotion', ha='center', va='center', rotation='vertical', fontsize=14, fontweight='bold')

    # add title to the figure
    fig.suptitle(title, fontsize=16, fontweight='bold')
    
    # plt.tight_layout()
    plt.show()
```
